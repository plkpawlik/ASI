LunarLander **DQN**

```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env

# Starting seed
SEED=69

# Setup enviroment
env = make_vec_env("LunarLander-v2", n_envs=4)
env.seed(SEED)

# Setup PPO model
model = PPO("MlpPolicy", env,
            seed=SEED,
            verbose=0)

# Train PPO  model
model.learn(total_timesteps=100_000)
model.save("LunarLander-v2-PPO")

# Clean up
del model
```

Save Videos

```python
import gnwrapper
from stable_baselines3 import SAC

# Setup enviroment
env = gnwrapper.Monitor(gym.make("LunarLander-v2", render_mode="rgb_array"), directory="./video/")

# Setup observation
observation = env.reset()
total_reward = 0

# Load PPO model
model=PPO.load("./LunarLander-v2-PPO")

while True:
    action, _states = model.predict(observation)
    observation, reward, done, info = env.step(action)
    total_reward += reward

    if done:
        env.close()
        env.reset()
        break

# Display videos
env.display()

# Log
print(f"Total reward: {total_reward}")
print(f"Saved videos: {env.videos}")
```
