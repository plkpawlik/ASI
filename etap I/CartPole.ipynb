CartPole **A2C**

```python
import gym
from stable_baselines3 import A2C
from stable_baselines3.common.env_util import make_vec_env

# Starting seed
SEED=69

# Setup enviroment
env = make_vec_env("CartPole-v1", n_envs=10)
env.seed(SEED)

# Setup A2C model
model = A2C("MlpPolicy", env,
            seed=SEED,
            gamma=0.95,
            learning_rate=0.01,
            ent_coef=0.01,
            vf_coef=0.75,
            verbose=0)

# Train A2C model
model.learn(total_timesteps=100_000)
model.save("./CartPole-v1-A2C")

# Clean up
del model
```

Save Videos

```python
import gnwrapper
from stable_baselines3 import A2C

# Setup enviroment
env = gnwrapper.Monitor(gym.make("CartPole-v1", render_mode="rgb_array"), directory="./video/")

# Setup observation
observation = env.reset()
total_reward = 0

# Load A2C model
model=A2C.load("./CartPole-v1-A2C")

while True:
    action, _states = model.predict(observation)
    observation, reward, done, info = env.step(action)
    total_reward += reward

    if done:
        env.close()
        env.reset()
        break

# Display videos
env.display()

# Log
print(f"Total reward: {total_reward}")
print(f"Saved videos: {env.videos}")
```
